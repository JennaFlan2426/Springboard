# Relax Take-Home Challenge
The brief: Defining  an  "adopted  user"   as  a  user  who   has  logged  into  the  product  on  three  separate days  in  at  least  one  seven-day  period ,  identify  which  factors  predict  future  user adoption.

```python
import numpy as np
import pandas as pd
users = pd.read_csv('takehome_users.csv', parse_dates=['creation_time'])
logins = pd.read_csv('takehome_user_engagement.csv', parse_dates=['time_stamp'])
```

```python
users['last_session_creation_time'] = pd.to_datetime(users['last_session_creation_time'], unit='s')
users.sample(10)
```

```python
logins.sample(10)
```

```python
logins.info()
```

```python
logins['day'] = pd.to_datetime(logins['time_stamp'].dt.date)
logins.info()
```

```python
roll_7d = logins.groupby('user_id').rolling('7D', on='day').count()
```

```python
roll_7d.head()
```

```python
actives = roll_7d[roll_7d['visited'] >= 3]
actives.head()
```

```python
activeids = actives.index.get_level_values(0).unique()
```

```python
users['active'] = np.where((users['object_id'].isin(activeids)),1,0)
users.head()
```

```python
users.info()
```

```python
users['invited_by_user_id'].unique()
```

```python
analysis = users.drop(['name', 'email'], axis=1)
analysis.info()
```

```python
ax1 = pd.get_dummies(data=analysis, columns=['creation_source'], prefix='source')
ax1.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
drop = ['object_id', 'creation_time', 'last_session_creation_time']
corr = ax1.drop(drop, axis=1).corr()
sns.heatmap(corr)
```

```python
corr['active']
```

Interesting. No strong correlations at all.

```python
for column in ax1.columns:
    print(column, ": ", ax1[column].nunique())
```

Good, all the yes/no columns have only 2 values. A question for later: it might be worth investigating if users that issue invitations are more likely to be active than others.

I think I will run this through decision tree classification since there are more yes/no columns than anything else.

```python
X = ax1.drop(drop, axis=1).fillna(0)
X.drop(['active'], axis=1, inplace=True)
y = ax1['active']
```

```python
from sklearn.tree import DecisionTreeClassifier 
from sklearn import tree
clf = DecisionTreeClassifier(random_state=42)
model = clf.fit(X, y)
fig = plt.figure(figsize=(25,20))
stuff = tree.plot_tree(clf, feature_names=X.columns, filled=True)
```

Okay, so that's not a decision tree, that's a decision Spanish moss. What are the feature importances?

```python
print('default (full) depth tree importances:')
for item in range(len(X.columns.values)):
    print(X.columns.values[item], ": ", clf.feature_importances_[item])
```

Let's see what a 5-depth tree looks like.

```python
clf = DecisionTreeClassifier(max_depth=5, random_state=42)
model = clf.fit(X, y)
fig = plt.figure(figsize=(25,20))
stuff5 = tree.plot_tree(clf, feature_names=X.columns, filled=True)
```

```python
text_representation = tree.export_text(clf, feature_names=X.columns.values.tolist())
print(text_representation)
```

```python
print('5-depth tree importances:')
for item in range(len(X.columns.values)):
    print(X.columns.values[item], ": ", clf.feature_importances_[item])
```

### Closing thoughts:
In both cases, 'org_id' is the most important feature. Interestingly, in the full-depth tree the next important feature is 'invited_by_user_id', where in the limited depth tree the "Personal Projects" source turns up as important.

This data does appear to be imbalanced. Many more users try the service than remain active users. For further investigation I would experiment with balancing the data via under- or over-sampling, or using bootstrap resampling to conduct multiple experiments during modeling. I would run Grid Search tuning of the model as well, to improve discovery.

In general, I would suggest that this data is missing features that might tell more of the story. Correlations were low across the board, possibly due to the majority of features being binary. Gini importance seems to focus on just one feature. Various human associations do seem to be strong factors; could we tease out additional information such as which particular organizations or users are connected with high proportions of active invitees? Are there user satisfaction metrics we could test?
